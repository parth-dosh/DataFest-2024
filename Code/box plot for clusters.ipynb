```python
#Relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
responses_df = pd.read_csv('responses.csv', low_memory=False)
```

```python
responses_df.head()
```

```python
print(responses_df.columns)
responses_df.describe()
```

```python
columns_to_keep = ['item_id', 'attempt', 'student_id']
respond_df = responses_df[columns_to_keep]
respond_df.head()
```

```python
respon_df = respond_df.groupby('student_id')['attempt'].mean().reset_index()
respon_df
```

```python
from matplotlib import pyplot as plt
respon_df['attempt'].plot(kind='line', figsize=(8, 4), title='attempt')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
respon_df = respon_df[respon_df['attempt']>1]
respon_df
```

```python
avg_attempt = respon_df['attempt'].mean()
avg_attempt
```

```python
# 25th percentile
quantile_25th = respond_df['attempt'].quantile(0.25)

# 75th percentile
quantile_75th = respon_df['attempt'].quantile(0.75)

# IQR
iqr = quantile_75th - quantile_25th

print("25th Percentile:", quantile_25th)
print("75th Percentile:", quantile_75th)
print("IQR:", iqr)
```

```python
quantile_75 = respon_df['attempt'].quantile(0.75)
respo_df = respon_df[respon_df['attempt']>quantile_75]
respo_df
```

```python
from matplotlib import pyplot as plt
respo_df['attempt'].plot(kind='line', figsize=(8, 4), title='attempt')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
#  iqr,  25th and 75th percentile for the 'attempt' column in respon_df

q1 = respo_df['attempt'].quantile(0.25)
q3 = respo_df['attempt'].quantile(0.75)
iqr = q3 - q1
print('25th percentile:', q1)
print('75th percentile:', q3)
print('IQR:', iqr)
```

```python
# dataframe respo_df: bar plot

respo_df.plot.bar(x='item_id', y='attempt')
```

```python
cluster_df = pd.read_csv('cluster_data.csv')
```

```python
# changing cluster_df dataframe to contain only those values that belong in cluster 2

cluster_2_df = cluster_df[cluster_df['Cluster'] == 2]
cluster_2_df.head()
```

```python
cluster_1_df = cluster_df[cluster_df['Cluster'] == 1]
cluster_1_df.head()
```

```python
# mean number of 'attempts' column for the items in respo_df taken from students in this (cluster_df) data frame (by cross-referencing 'student_id' in both)

import pandas as pd
merged_df_2 = pd.merge(respo_df, cluster_2_df, on='student_id', how='inner')
mean_attempts_2 = merged_df_2['attempt'].mean()
print(f"Mean number of attempts for items in respo_df taken from students in cluster_df: {mean_attempts_2}")
```

```python
#mean number of 'attempts' column for the items in respo_df taken from students in tis (cluster_df) data frame (by cross-referencing 'student_id' in both)

import pandas as pd
merged_df_1 = pd.merge(respo_df, cluster_1_df, on='student_id', how='inner')
mean_attempts_1 = merged_df_1['attempt'].mean()
print(f"Mean number of attempts for items in respo_df taken from students in cluster_df: {mean_attempts_1}")
```

```python
cluster_0_df = cluster_df[cluster_df['Cluster'] == 0]
cluster_0_df.head()
```

```python
# mean number of 'attempts' column for the items in respo_df taken from students in this (cluster_df) data frame (by cross-referencing 'student_id' in both)

import pandas as pd
merged_df_0 = pd.merge(respo_df, cluster_0_df, on='student_id', how='inner')
mean_attempts_0 = merged_df_0['attempt'].mean()
print(f"Mean number of attempts for items in respo_df taken from students in cluster_df: {mean_attempts_0}")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# boxplot for each merged dataframe
sns.boxplot(x="Cluster", y="attempt", data=pd.concat([merged_df_0, merged_df_1, merged_df_2]))

#title and labels
plt.title("Comparison of Attempts by Cluster")
plt.xlabel("Cluster")
plt.ylabel("Attempts")

plt.show()
```
